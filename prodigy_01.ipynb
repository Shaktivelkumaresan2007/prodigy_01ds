```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/population.csv")
```

```python
df.head(266)
```

```python
df.describe()
```

```python
countries_to_plot = ['United States','Zambia','Canada', 'Brazil', 'Japan', 'Germany','Afghanistan','France']
filtered_data = df[df['Country Name'].isin(countries_to_plot)]
plt.bar (filtered_data['Country Name'], filtered_data['2022'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2022')
plt.title('Population of selected countries in 2022')
plt.xticks(rotation=45, ha='right')
plt.show()
```
